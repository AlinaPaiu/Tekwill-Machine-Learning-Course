**Scrieți o funcție care să calculeze volumul unei sfere, dată fiind raza acesteia.**

<p>Volumul unei sfere este dat de formula $$\frac{4}{3} πr^3$$</p>

```python
def volumul(r):
  pi = 3.14

  return round(4/3.0*pi*r**3,3) #punem 0 ca rezultatul sa fie float
```

```python
volumul(1)
```

**Scrieți o funcție care să verifice dacă un număr se află într-un interval dat (inclusiv valorile mare și mică)**

```python
def este_in_interval(start,end,nr):
  interval = [digit for digit in range(start,end+1)]
  if nr in interval:
    print(f'Numarul {nr} este in intervalul {start}-{end}')
  else:
    print(f'Numarul {nr} nu este in intervalul {start}-{end}')
```

```python
este_in_interval(1,10,10)
```

**Scrieți o funcție Python care acceptă un șir de caractere și calculează numărul de litere majuscule și minuscule.**

    Exemplu de String : 'Hello Mr. Rogers, how are you this fine Tuesday?'
    Rezulat asteptat : 
    Nr de majuscule : 4
    Nr de minuscule : 33

SUGESTIE: Două metode de string care s-ar putea dovedi utile: **.isupper()** and **.islower()**

```python
def func3(string):
    upper = 0
    lower = 0

    for letter in string:
      if letter.isupper():
        upper += 1
      else: lower +=1
    print(f'String: {string}')
    print(f'Nr de majuscule: {upper}')
    print(f'Nr de minuscule: {lower}')
```

```python
func3('Today is a Great Day!')
```

**Scrieți o funcție Python care ia o listă și returnează o nouă listă cu elemente unice ale primei liste.**

    Sample List : [1,1,1,1,2,2,3,3,3,3,4,5]
    Unique List : [1, 2, 3, 4, 5]

```python
def func4(list):
  my_set = set(list)
  set_to_list = sorted(my_set)
  print(f'Lista initiala (cu elemente duplicate): {list}')
  print(f'Lista cu elemente unice: {set_to_list}')
```

```python
func4([1,1,1,1,2,2,3,3,3,3,4,5])
```

**Scrieți o funcție Python care să înmulțească toate numerele dintr-o listă.**

    Sample List : [1, 2, 3, -4]
    Expected Output : -24

```python
def func5(list):
  result = 1
  for element in list:
      result = result * element
  return result
```

```python
func5([1,2,3,-4])
```

**Scrieți o funcție Python care să verifice dacă un șir de caractere transmis este sau nu palindrom.**

Notă: Un palindrom este un cuvânt, o frază sau o secvență care se citește la fel si în sens invers, de exemplu, madam sau ana.

```python
def func6(str):
  lower_str = str.lower()
  if lower_str == lower_str[::-1]:
    print(f"Sirul de caractere '{str}' ESTE un polindrom")
  else: print(f"Sirul de caractere '{str}' NU ESTE un polindrom")
```

```python
func6('Madam')
```

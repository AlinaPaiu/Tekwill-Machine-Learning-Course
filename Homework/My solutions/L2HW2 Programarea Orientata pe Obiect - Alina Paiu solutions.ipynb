#### Problema 1
Completați metodele clasei Linie pentru a accepta coordonatele ca o pereche de tupluri și pentru a returna panta și distanța liniei.


```python
import math

class Line:
    
    def __init__(self,coor1,coor2):
        self.coor1 = coor1
        self.coor2 = coor2
    
    def distance(self):
      
        return round(math.sqrt((self.coor2[0]-self.coor1[0])**2 + (self.coor2[1]-self.coor1[1])**2),2)

    def slope(self):
        return round((self.coor2[1]-self.coor1[1])/(self.coor2[0]-self.coor1[0]),2)
```

```python
# EXAMPLE OUTPUT

coordinate1 = (16,25)
coordinate2 = (4,5)

li = Line(coordinate1,coordinate2)
```

```python
li.distance()
```

```python
li.slope()
```

```python
import math

class Cylinder:
    
    def __init__(self,height=1,radius=1):
        self.height = height
        self.radius = radius
        
    def volume(self):

        return math.pi*self.height*(self.radius)**2
    
    def surface_area(self):
       
        return 2*math.pi*self.radius*(self.radius+self.height)
```

```python
# EXAMPLE OUTPUT
c = Cylinder(2,3)
```

```python
c.volume()
```

```python
c.surface_area()
```
